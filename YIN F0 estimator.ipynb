```python
t0 = 8000 + 10000
W = int( 25e-3 * 50000 ) # 5000
corr = np.correlate(psub[t0:t0 + 2 * W - 1], psub[t0:t0 + W], mode = 'valid')
```

```python
print W
print len(corr)
```

```python
plt.close('all')
plt.plot(corr, 'k')
plt.show()
```

```python
diff = np.zeros(W)

for tau in range(W):
    diff[tau] = np.sum( (psub[t0:t0 + W] - psub[t0 + tau:t0 + W + tau])**2 )
```

```python
plt.close('all')

plt.plot(diff, 'k')

plt.show()
```
